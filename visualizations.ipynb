```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path = 'C:/Users/14861/Desktop/新数科概（修改配色版）/heart_statlog_cleveland_hungary_final.csv'
data = pd.read_csv(file_path)

sns.set(style='whitegrid')
```

```python
#绘制年龄分布图
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=30, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

for item in data['resting bp s']:
    if item < 100:
        print(item)
```

```python

#绘制静息血压与胆固醇水平比较散点图
plt.figure(figsize = (10, 6))
sns.scatterplot(x= 'resting bp s', y = 'cholesterol', hue = 'target', data = data)
plt.title('Resting blood pressure and Cholesterol levels')
plt.xlabel('Resting blood pressure')
plt.ylabel('Cholesterol')
plt.legend(title='Heart disease')
plt.show()
```

```python

#绘制胸痛程度分部图
plt.figure(figsize=(10, 6))
sns.countplot(x = 'chest pain type', data = data)
plt.title('Distribution of chest pain types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python

#运动心绞痛与最大心率箱线图
plt.figure(figsize=(10, 6))
sns.boxplot(x='exercise angina', y='max heart rate', data=data, palette='muted')
plt.title('Maximum heart rate by exercise induced angina')
plt.xlabel('Exercise induced angina')
plt.ylabel('Maximum heart rate')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python

#相关性热力图
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='vlag', fmt=".2f")
plt.title('Correlation heatmap')
plt.xticks(rotation = -30)
plt.show()
```

```python
data_new = data.copy()
data_new['target'] = data['target'].replace({1:'Heart Disease', 0:'Normal'})
data_new['sex'] = data['sex'].replace({1:'Male', 0:'Female'})

#最大心率与年龄
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='max heart rate', hue='target', data=data_new)
plt.title('Maximum heart rate vs age')
plt.xlabel('Age')
plt.ylabel('Maximum heart rate')
plt.legend(title='Heart disease')
plt.show()
```

```python
# 每个性别的患病比例
counts = data_new.groupby(['sex', 'target']).size().unstack()
totals = counts.sum(axis=1)
percentages = counts.divide(totals, axis=0) * 100

percentages.plot(kind='bar', stacked=True, colormap='tab20')

plt.xlabel('Sex')
plt.ylabel('Percentage')
plt.title('Stacked Bar Plot of Sex and Heart Disease')
plt.xticks(rotation=0)
plt.legend(title='Heart Disease')
plt.show()
```
